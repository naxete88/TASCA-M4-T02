## Histograms

Dot plots, like in scatterplot, show the exact value for each observation. This is useful for small datasets, but they can become hard to read with larger samples.

Rather than showing the value of each observation, we prefer to think of the value as belonging to a bin.

These bins - (counts) are plotted as bars into what is called a Histogram.

Histogram provide a view of the data density. Higher bars represent where the data are relatively more common.

Histogram are especially convenient for describing the shape of the data distribution.

When data trail off to the right and have a longer right tail, the shape is said to be Right Skewed or also called Skewed to the Positive End.   
Contrary, data with the reverse characteristic – a long, thin tail to the left – are said to be Left Skewed. We also say that such a distribution has a long left tail.   
Data that show roughly equal trailing off in both directions are called Symmetric.

When data trail off in one direction, the distribution has a long tail. If a distribution has a long left tail, it is Left Skewed. If a distribution has a long right tail, it is Right Skewed.

Fes les tasques de preprocessat  i adequació del Dataset que disposem en el repositori de GitHub PRE-PROCESSING-DATA amb l'objectiu de preparar-lo i treballar-lo com a dataframe per a extreure'n informació. El resultat final ha de ser un dataframe amb les variables de nom de la pel·lícula, any i genere. També és important quedar-se amb un sol gènere per a cada pel·lícula!

```python
df_movies = pd.read_csv("https://raw.githubusercontent.com/BIDDL3/PRE-PROCESSING-DATA/main/movies.dat",
                       sep='::',
                       encoding = 'ISO-8859-1',
                       index_col = 0,
                       names = ['ID','Nom','Genere','Any'],
                       engine = 'python',
                       )
# extrec l'any de la columna nom 
df_movies['Any'] = df_movies['Nom'].str[-5:-1].astype(int)
# un cop tinc la columna d'any feta elimino els anys de la columna nom
df_movies['Nom'] = df_movies['Nom'].str[:-6]
# les coloco en l'ordre correcte
df_movies = df_movies.reindex(columns=['Nom','Any','Genere'])
# trio el primer génere abans del primer separador perque només tinguem un génere:
df_movies['Genere'] = df_movies['Genere'].str.split("|").str[0]
# demano que em mostri els primers registres 
df_movies.head(10)
```

### Mostra la teva creativitat. Què creus rellevant mostrar del Dataset "movies.dat" de l'exercici anterior? Fes una o dues representacions gràfiques i justifica la teva elecció.

Crec que és important veure com s'han distribuit els generes al llarg dels anys. Per aquest motiu he utilitzat un sistema catplot de violí, tenint per un costat una variable numérica com és l'any i després una variable categòrica com són els gèneres, per una banda he fet un gràfic separant cada gènere i després he fet un gràfic unint tots els gèneres, per una banda puc veure com és el comportament de cada gènere de forma individualitzada i després per veure com es comporten tots d'una manera més generalitzada.

```python
ax = sns.catplot(x='Any', 
                 y='Genere', 
                 kind='violin',
                 data=df_movies,
                )
ax.fig.set_figwidth(20)
ax.fig.set_figheight(80)
```

Amb aquest gràfic podem veure coses interessants:
- Animació: comença la producció a partir dels anys 20, i augmenta en gran mesura a partir dels 80.
- Adventures: comença la producció cap a 1910, i arriba al seu màxim cap a 1980 - 2000.
- Comedia, Acció i Drama: aquests 3 géneres tenen un comportament molt similar, comencen cap a 1910 i porten una producció baixa fins a arribar als anys 80-90. He ajuntat aquests tres gèneres perque segueixen una tendència molt semblant.
- Thriller: comença la producció cap a 1910 i augmenta als anys 80-90
- Crime: Comença una producció molt baixa, augmenta lleugerament a partir dels 40 i augmenta a partir dels 80-90.
- Romance
- Infantil: producció tardana cap als anys 30, i comença a augmentar amb més força a partir dels anys 60.
- Documental: la seva producció comença a partir de 1920 i té la seva màxima producció entre els 90 i els 2000.
- Sci-fi: comença sobre 1910, a partir de 1940 augmenta i manté una producció constant fins els anys 2000 on baixa amb més força
- Horror: comença a 1910 i arriba al seu màxim entre 1980 i 2000.
- Western: comença a finals de 1910, augmenta fins arribar al seu màxim sobre els 70 i després comença a baixar fins a 2010.
- Misteri: comença sobre 1910, augmenta sobre 1940 i augmenta encara més entre els 80 i els 2000.
- Film-Noir: comença a partir de 1920, té un augment entre 1940 i 1960, i té un repunt entre els 80 i els 2000.
- War: el génere està molt repartit i de fet la producció per anys es breu però continua. Curiosament augmenta lleugerament a partir dels 40, quan acaba la segona guerra mundial i ocupa part de la guerra freda.
- Fantasia: té una concentració molt gran entre els anys 80 i 90.
- Musicals: la producció és més grans entre els anys 40 i 60, als anys 70 comencen a baixar i ja entre 2000 i 2020 no hi ha dades.
